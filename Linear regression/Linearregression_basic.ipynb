```python
#loading libraries
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create an instance of the LinearRegression model
line_fitter = LinearRegression()
```

We have the temperature and ice creame sales data. We are going to test the linear regression model on temperature verses sales.

```python
# Create input values
temperature = np.array(range(60, 100, 2))
temperature = temperature.reshape(-1, 1)
sales = [65, 58, 46, 45, 44, 42, 40, 40, 36, \
				38, 38, 28, 30, 22, 27, 25, 25, 20, 15, 5]
```

```python
# Fit the Linear Regression model using temperature and sales data
line_fitter.fit(temperature, sales)

# Predict sales using the fitted model
sales_predict = line_fitter.predict(temperature)
```

```python
# Plot the original sales data points as circles ('o')
plt.plot(temperature, sales, 'o', label='Actual Sales')
# Plot the predicted sales data points as a line
plt.plot(temperature, sales_predict, label='Predicted Sales')
# Add labels and legend to the plot
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.title('Relationship between Temperature and Sales')
plt.legend()
# Show the plot
plt.show()
```
